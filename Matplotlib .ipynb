Matplotlib est une bibliothèque python qui dessine des graphiques . Nul besoin de connaissances en interfaces graphiques pour créer un graphique dynamique avec possibilité de zoom et de sauvegarde par l'utilisateur. Il est d'ailleurs possible de sauvegarder les graphiques en format matriciels comme le PNG, JPEG , etc. et vectoriels comme le PDF et le SVG . Une grande communauté s'est formée autour de ce projet et donc la documentation et les exemples ne manquent pas. 

```python
import matplotlib.pyplot as plt

plt.title("Danger de la vitesse")
x = [50,80,90,110,115]
y = [1,2,3,4,5]
plt.plot(x, y)
plt.xlabel('Vitesse')
plt.ylabel('Temps')
plt.show()
```
