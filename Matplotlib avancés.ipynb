```python
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0,2*np.pi,50)
x = np.sin(t)
y = np.cos(t)
plt.figure()     # Créer une figure
plt.plot(x)      # Tracer X dans cette figure 
plt.title("sin") # le titre de la figure 
plt.figure()     # Créer une autre  figure
plt.plot(y)      # Tracer Y dans cette figure 
plt.title("Cos") # le titre de la figure 
```

### <b> <span style="color:blue"> Scatter plots 

```python
from numpy.random import rand

N = 30 # nombre de point à tracer 
x = np.linspace(0,10, N) 
e1 = rand(N) * 0.8
y1 = x + e1
areas = rand(N)*600
plt.figure(figsize=(12,5))  # créer une figure avec  largeur =12 inches et hauteur = 5 inches .
########################
plt.subplot(1,2,1) 
# plt.subplot() dessinez plusieurs tracés sur une seule figure, 
# Vous pouvez dessiner autant de tracés que vous le souhaitez sur une seule figure, 
# il suffit de décrire le nombre de lignes, de colonnes et l'index du tracé.
plt.scatter(x, y1, s=areas)
colors = rand(N)
plt.scatter(x, y1, s=areas ,c=colors)
plt.colorbar() # traçcer la barre de couleurs 
plt.title("Random scatter")

e2 = rand(N) * 0.15
y2 = x + e2
areas2 = rand(N)* 1500
########################
plt.subplot(1,2,2)
colors = rand(N)
plt.scatter(x, y2, s=areas2 ,c=colors)
plt.colorbar() # traçcer la barre de couleurs 
plt.title("Random scatter 2")
```

### <b> <span style="color:blue">  Matrice de corrélation

```python
#génération des données 
e1 = rand(100) # générer 100 valeure entre 0 et 1 
e2 = rand(100)*2
e3 = rand(100)*10
e4 = rand(100)*100

# Calculer le coefficient de corrélation de Pearson à l'aide de np.corrcoef
matrix_de_correlation   = np.corrcoef([e1, e2,e3, e4]) 
# plt.imshow() pour Visualiser la matrice de corrélation
plt.imshow(matrix_de_correlation,cmap='Reds') # https://matplotlib.org/stable/tutorials/colors/colormaps.html
plt.colorbar()
```

### <b> <span style="color:blue"> Histogram

```python
from numpy.random import randint
data = randint(100, size=(10,50)) #Renvoie des entiers aléatoires 
data2 = randint(50, size=(10,50))
x = np.sum(data, axis=0)
x2 = np.sum(data2, axis=0)

plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.hist(x, color='r', label="X")
plt.title("hist")
plt.legend()

plt.subplot(2,2,2)
plt.hist(x, cumulative=True, label="cumulative")
plt.title("cumulative=True")
plt.legend()

plt.subplot(2,2,3)
plt.hist([x, x2], label=["x", "x2"]) #  pour  un histograms multiple :
plt.title("histograms multiple")
plt.legend()
```
